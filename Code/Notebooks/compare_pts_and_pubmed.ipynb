# Read in the pubmed curated dataframe

```python
pubmed_df = pd.read_csv('../Results/tempus.tsv', sep='\t')
print(pubmed_df.shape)
pubmed_df.head()
```

# Read in the PTS archive sheet

```python
pts_df = pd.read_csv('../Data/PTS_archive_03_25_22.tsv', sep='\t', skiprows=1)
print(pts_df.shape)
print(pts_df.columns)
pts_df.head()
```

```python
print(pts_df.shape)
print(pts_df['Publication Type'].value_counts())
pts_df = pts_df[pts_df['Publication Type']=='manuscript']
print()
print(pts_df.shape)
print(pts_df['Authorship Lead'].value_counts())
pts_df = pts_df[(pts_df['Authorship Lead']=='Collab 1st and Last Author (Tempus Authored)')\
                  | (pts_df['Authorship Lead'] == 'Tempus 1st &/or Last Author')]

print(pts_df.shape)
for bad_journal in ['arxiv', 'biorxiv', 'medrxiv']:
    pts_df = pts_df[pts_df['Journal / Conference'].str.contains(bad_journal, case=False) == False]
    print(pts_df.shape)
```

```python
pts_df[pts_df['PMID'].isnull()==True][['Title/Topic/Link', 'Journal / Conference']]
```

```python
print(pts_df.shape)
pts_df = pts_df[pts_df['PMID'].isnull()==False]
print(pts_df.shape)
pts_df['PMID'] = pts_df['PMID'].astype('int64')###Just changing the "type" of the column, python detail
```

# Merge PTS and pubmed dataframes to look for (and/or categorize) inconsistencies

```python
print('Number of PMIDs in the PTS that are not in the pubmed df:',
                          len(set(pts_df['PMID'])-set(pubmed_df['PMID'])))

pts_only_df = pts_df[pts_df['PMID'].isin(pubmed_df['PMID'])==False]
pts_only_df
```

```python
print('Number of PMIDs in the pubmed df that are not in the PTS:',
                              len(set(pubmed_df['PMID'])-set(pts_df['PMID'])))
pubmed_only_df = pubmed_df[pubmed_df['PMID'].isin(pts_df['PMID'])==False]
pubmed_only_df
```

```python
print('Only those with more than one Tempus author:')
pubmed_df[(pubmed_df['PMID'].isin(pts_df['PMID'])==False) & (pubmed_df['Affiliated_author_count']>1)]
```
