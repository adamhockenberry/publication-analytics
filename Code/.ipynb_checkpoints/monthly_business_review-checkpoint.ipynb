# Reading in the `CSV`

```python
# df = pd.read_csv('../Data/PTS_05_25_21.tsv', sep='\t', header=1)
df = pd.read_csv('../Data/PTS_05_27_21.tsv', sep='\t', header=1)
df.head()
```

**Just printing the columns to keep track of them**

```python
print(df.columns)
```

**Missing values can cause some bugs so fill the `NaN`s with empty strings for these date columns**

```python
cols_to_fill = ['Date of Most Recent Resubmission (Yr, Quarter)',
                'Date Actual Submission Occurred (Yr, Quarter, Date)',
                'Date of Publication (Data / Yr, Quarter)']
for col in cols_to_fill:
    df[col] = df[col].fillna('')
```

# Making a minimal `dataframe`

**Not sure if this is the best way to do it but seems to be a good start. The full table contains way more than we need for these initial analyses.**

```python
###Chooose the quarters we want to care about
# quarters_of_interest = ['2020 Q1', '2020 Q2', '2020 Q3', '2020 Q4', '2021 Q1', '2021 Q2']
# quarters_of_interest = ['2020 Q1', '2020 Q2', '2020 Q3', '2020 Q4']
quarters_of_interest = ['2021 Q1', '2021 Q2']
```

```python
###Copy out the columns that we care about (and only those columns)
tiny_df = df[['Title/Topic/Link',
              'Origin (internal vs. external)',
              'Publication Type',
              'Key BU Stakeholder Publication Supports']].copy()
tiny_df.rename(columns={'Title/Topic/Link': 'Title',
                        'Origin (internal vs. external)': 'Origin',
                        'Key BU Stakeholder Publication Supports': 'Stakeholder'}, inplace=True)
tiny_df.head()
```

```python
tiny_df['Origin'].value_counts()
```

```python
print(tiny_df.shape)
tiny_df = tiny_df[tiny_df['Origin'].isin(['Tempus led', 'Collaborator led (Tempus authorship)'])]
print(tiny_df.shape)
```

```python
tiny_df['Publication Type'].value_counts()
```

```python
print(tiny_df.shape)
tiny_df = tiny_df[tiny_df['Publication Type'].isin(['abstract','manuscript'])]
print(tiny_df.shape)
```

```python
###Make some clean columns
tiny_df['Submission_status'] = ''
tiny_df['Publication_status'] = ''
tiny_df['Submission_quarter'] = ''
tiny_df['Publication_quarter'] = ''

###Iterate throught the quarters that I care about
for quarter in quarters_of_interest:
    print(quarter)
    ###################################################################
    ###Select the lines that were published this quarter
    temp_df = df[df['Date of Publication (Data / Yr, Quarter)'].isnull()==False]
    temp_df = temp_df[(temp_df['Date of Publication (Data / Yr, Quarter)'].str.contains(quarter))]
    indices = [i for i in temp_df.index if i in tiny_df.index]
    ###And set values in my tiny_df
    tiny_df.at[indices, 'Publication_status'] = 'Published'
    tiny_df.at[indices, 'Publication_quarter'] = quarter
    ###################################################################
    ###Now select the lines that were submitted (or re-submitted) this quarter
    temp_df = df[(df['Date of Most Recent Resubmission (Yr, Quarter)'].str.contains(quarter)) 
                | (df['Date Actual Submission Occurred (Yr, Quarter, Date)'].str.contains(quarter))]
    indices = [i for i in temp_df.index if i in tiny_df.index]
    ###And update these values as well
    tiny_df.at[indices, 'Submission_status'] = 'Submitted'
    tiny_df.at[indices, 'Submission_quarter'] = quarter
    
###Limit my tiny dataframe to only care about projects that were either submitted or published
tiny_df = tiny_df[(tiny_df['Submission_status']=='Submitted') | (tiny_df['Publication_status']=='Published')]
print(tiny_df.shape)
tiny_df.head()
```

# Stakeholders....

```python
from cycler import cycler
```

```python
colors = ['#39ABD1', '#4651C3', '#836CD0', '#A8CE40', '#36BC7D',\
          '#EFC72A', '#AE1860', '#F76719', '#8C8C8C', '#C7C7C7']

colors = ['#288CAE', '#39ABD1', '#62BCDA', '#8BCEE4', '#B5DFEE', '#DEF1F7',
          '#B4221D','#DD2E27', '#E45853', '#EB837F', '#F2AEAB', '#F9D9D7']
custom_cycler = (cycler(color=colors))
```

```python
possible_stakeholders = tiny_df['Stakeholder'].value_counts().index
```

```python
x_labels = [i.split(' ')[-1] for i in quarters_of_interest]
print(x_labels)
x = np.arange(len(x_labels))
print(x)

dicty = {}
for quarter in quarters_of_interest:
    dicty[quarter] = {}
    for pubtype in ['abstract', 'manuscript']:
        dicty[quarter][pubtype] = {}
        tempy_df = tiny_df[(tiny_df['Submission_status']== 'Submitted')&(tiny_df['Submission_quarter']==quarter)]
        tempy_df = tempy_df[tempy_df['Publication Type']==pubtype]
        for i in possible_stakeholders:
            dicty[quarter][pubtype][i] = tempy_df[tempy_df['Stakeholder']==i].shape[0]

width = 0.35

fig, ax = plt.subplots()
ax.set_prop_cycle(custom_cycler)
quarter_index = 0.5
for quarter in quarters_of_interest:
    x_modifier = -width/2
    for pubtype in ['abstract', 'manuscript']:
        bottom = 0
        for sh in possible_stakeholders[::-1]:
            barheight = dicty[quarter][pubtype][sh]
            if quarter == quarters_of_interest[0]:
                tempy = ax.bar(quarter_index+x_modifier, barheight,
                           width, bottom=bottom,
                           label=sh + ' ' + pubtype)
            else:
                tempy = ax.bar(quarter_index+x_modifier, barheight,
                           width, bottom=bottom)
            bottom += barheight
        x_modifier+=width
    quarter_index += 1
    
ax.set_ylabel('Counts')
ax.set_xticks([i+0.5 for i in range(len(quarters_of_interest))])
ax.set_xticklabels(x_labels)
ax.set_title('Submitted')
handles, labels = ax.get_legend_handles_labels()
handles = handles[:6][::-1]+handles[6:][::-1]
labels = labels[:6][::-1]+labels[6:][::-1]
# # sort both labels and handles by labels
# labels, handles = zip(*sorted(zip(labels, handles), key=lambda t: t[0]))
# ax.legend(handles, labels)
ax.legend(handles, labels, ncol=2, loc='upper center', bbox_to_anchor=(0.5, -0.2))
ax.yaxis.set_major_locator(MaxNLocator(integer=True))
# plt.savefig('../submitted_2020and21_noN.pdf', bbox_inches='tight')
```

# Getting numbers and titles using pivot tables

```python
pivoted_df = pd.pivot_table(tiny_df[tiny_df['Submission_status']=='Submitted'], 
               values=['Submission_status'], 
               index=['Publication Type', 'Submission_quarter'],
               columns=['Stakeholder'], aggfunc='count')
pivoted_df = pivoted_df.fillna(0)
for column in pivoted_df.columns:
    pivoted_df[column] = pivoted_df[column].astype(int)
    
pivoted_df.to_csv('../stakeholder_2020and2021_submission_summary.csv')

pivoted_df = pd.pivot_table(tiny_df[tiny_df['Publication_status']=='Published'], 
               values=['Publication_status'], 
               index=['Publication Type', 'Publication_quarter'],
               columns=['Stakeholder'], aggfunc='count')
pivoted_df = pivoted_df.fillna(0)
for column in pivoted_df.columns:
    pivoted_df[column] = pivoted_df[column].astype(int)

pivoted_df.to_csv('../stakeholder_2020and2021_publication_summary.csv')
```

```python
temp_df = tiny_df[tiny_df['Submission_status']=='Submitted'][['Publication Type', 'Submission_quarter', 'Title', 'Stakeholder']]
temp_df = temp_df.set_index(['Publication Type', 'Submission_quarter', 'Title']).sort_index()
temp_df.to_csv('../submitted_stakeholder_titles.csv')
temp_df = tiny_df[tiny_df['Publication_status']=='Published'][['Publication Type', 'Publication_quarter', 'Title', 'Stakeholder']]
temp_df = temp_df.set_index(['Publication Type', 'Publication_quarter', 'Title']).sort_index()
temp_df.to_csv('../published_stakeholder_titles.csv')
```
